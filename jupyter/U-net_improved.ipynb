## Data

```python
import os
import numpy as np
from skimage.io import imsave, imread
from tqdm import tqdm_notebook
from matplotlib import pyplot as plt
%matplotlib inline
```

The original images are 420x580.

```python
data_path = '../raw/'

image_rows = 420
image_cols = 580
```

### Load the training images

```python
train_data_path = os.path.join(data_path, 'train')
images = os.listdir(train_data_path)
total = len(images) // 2
```

```python
train_imgs = np.ndarray((total, image_rows, image_cols), dtype=np.uint8)
train_imgs_mask = np.ndarray((total, image_rows, image_cols), dtype=np.uint8)

i = 0
for image_name in tqdm_notebook(images):    
    if 'mask' in image_name:
        continue
        
    image_mask_name = image_name.split('.')[0] + '_mask.tif'
    img = imread(os.path.join(train_data_path, image_name), as_grey=True)
    img_mask = imread(os.path.join(train_data_path, image_mask_name), as_grey=True)

    img = np.array([img])
    img_mask = np.array([img_mask])

    train_imgs[i] = img
    train_imgs_mask[i] = img_mask
    i += 1
```

Save/load to save time when rerunning this notebook.

```python
np.save('imgs_train.npy', train_imgs)
np.save('imgs_mask_train.npy', train_imgs_mask)
```

```python
train_imgs = np.load('imgs_train.npy')
train_imgs_mask = np.load('imgs_mask_train.npy')
```

#### Let's look at some of the training images and masks

```python
def plot_n(datasets, headers, n, width=10, height=40):
    """Plot the images alongside their masks
    
    datasets -- a tuple of numpy ndarrays
    (e.g., one array for the original images,
    one array for true masks, and one array for the predicted masks)
    headers -- a tuple containing a header for each dataset
    n -- how many sets of images-with-masks to plot
    """
    
    n_cols = len(datasets)
    f = plt.figure(figsize=(width, height))
    
    for i in range(n):
        j = 1
        for header, imgs in zip(headers, datasets):
            sp = f.add_subplot(n, n_cols, n_cols*i+j)
            sp.axis('Off')
            sp.set_title(header)
            imgs = np.squeeze(imgs)
            plt.imshow(imgs[i, :, :], cmap="gray")
            j += 1
```

### Train-validation split

```python
np.random.seed(2018)
train_size = int(0.8 * train_imgs.shape[0])
train_ind = np.random.choice(train_imgs.shape[0], size=train_size, replace=False)
train_split = np.zeros(train_imgs.shape[0])
train_split[train_ind] = 1

imgs_train = train_imgs[train_split == 1, :, :]
imgs_val = train_imgs[train_split == 0, :, :]
imgs_mask_train = train_imgs_mask[train_split == 1, :, :]
imgs_mask_val = train_imgs_mask[train_split == 0, :, :]

print('Images:')
print('Full data: {}'.format(train_imgs.shape))
print('Training set: {}'.format(imgs_train.shape))
print('Validation set: {}'.format(imgs_val.shape))
print('Masks:')
print('Full data: {}'.format(train_imgs_mask.shape))
print('Training set: {}'.format(imgs_mask_train.shape))
print('Validation set: {}'.format(imgs_mask_val.shape))
```

## Training a deep neural network

```python
from skimage.transform import resize
from keras.models import Model
from keras.layers import Input, concatenate, Conv2D, MaxPooling2D, Conv2DTranspose
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint
from keras import backend as K

K.set_image_data_format('channels_last')  # TF dimension ordering in this code
```

Evaluation metric is the Dice coefficient.

```python
def dice_coef(y_true, y_pred, smooth=1.0):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2.0 * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)


def dice_coef_loss(y_true, y_pred):
    return -dice_coef(y_true, y_pred)
```

We resize, center, and scale the images before training.

```python
resized_img_rows = 80
resized_img_cols = 112

def preprocess(imgs, img_rows=None, img_cols=None):
    imgs_p = np.ndarray((imgs.shape[0], img_rows, img_cols), dtype=np.uint8)
    for i in range(imgs.shape[0]):
        imgs_p[i] = resize(imgs[i], (img_rows, img_cols), preserve_range=True)

    imgs_p = imgs_p[..., np.newaxis]
    return imgs_p


imgs_train = preprocess(imgs_train, img_rows=resized_img_rows, img_cols=resized_img_cols)
imgs_mask_train = preprocess(imgs_mask_train, img_rows=resized_img_rows, img_cols=resized_img_cols)
imgs_val = preprocess(imgs_val, img_rows=resized_img_rows, img_cols=resized_img_cols)
imgs_mask_val = preprocess(imgs_mask_val, img_rows=resized_img_rows, img_cols=resized_img_cols)

imgs_train = imgs_train.astype('float32')
imgs_val = imgs_val.astype('float32')

mean_train = np.mean(imgs_train)  # mean for data centering
std_train = np.std(imgs_train)  # std for data normalization

imgs_train -= mean_train
imgs_train /= std_train
imgs_mask_train = imgs_mask_train.astype('float32')
imgs_mask_train /= 255.  # scale masks to [0, 1]

imgs_val -= mean_train
imgs_val /= std_train
imgs_mask_val = imgs_mask_val.astype('float32')
imgs_mask_val /= 255.  # scale masks to [0, 1]

print('Images:')
print('Full data: {}'.format(train_imgs.shape))
print('Training set: {}'.format(imgs_train.shape))
print('Validation set: {}'.format(imgs_val.shape))
print('Masks:')
print('Full data: {}'.format(train_imgs_mask.shape))
print('Training set: {}'.format(imgs_mask_train.shape))
print('Validation set: {}'.format(imgs_mask_val.shape))
```

### U-Net

![U-net diagram](../img/u-net-architecture.png)

```python
inputs = Input((resized_img_rows, resized_img_cols, 1))
conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv1)
pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(pool1)
conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv2)
pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')(pool2)
conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')(conv3)
pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)

conv4 = Conv2D(256, (3, 3), activation='relu', padding='same')(pool3)
conv4 = Conv2D(256, (3, 3), activation='relu', padding='same')(conv4)
pool4 = MaxPooling2D(pool_size=(2, 2))(conv4)

conv5 = Conv2D(512, (3, 3), activation='relu', padding='same')(pool4)
conv5 = Conv2D(512, (3, 3), activation='relu', padding='same')(conv5)

up6 = concatenate([Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(conv5), conv4], axis=3)
conv6 = Conv2D(256, (3, 3), activation='relu', padding='same')(up6)
conv6 = Conv2D(256, (3, 3), activation='relu', padding='same')(conv6)

up7 = concatenate([Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(conv6), conv3], axis=3)
conv7 = Conv2D(128, (3, 3), activation='relu', padding='same')(up7)
conv7 = Conv2D(128, (3, 3), activation='relu', padding='same')(conv7)

up8 = concatenate([Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(conv7), conv2], axis=3)
conv8 = Conv2D(64, (3, 3), activation='relu', padding='same')(up8)
conv8 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv8)

up9 = concatenate([Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(conv8), conv1], axis=3)
conv9 = Conv2D(32, (3, 3), activation='relu', padding='same')(up9)
conv9 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv9)

conv10 = Conv2D(1, (1, 1), activation='sigmoid')(conv9)

model = Model(inputs=[inputs], outputs=[conv10])

model.summary()
```

```python
model.compile(optimizer=Adam(lr=1e-5), loss=dice_coef_loss)

model_checkpoint = ModelCheckpoint('weights.h5', monitor='val_loss', save_best_only=True)

model.fit(imgs_train, imgs_mask_train, batch_size=32, epochs=50, verbose=1, shuffle=True,
          validation_data=(imgs_val, imgs_mask_val), callbacks=[model_checkpoint])
```

```python
model.compile(optimizer=Adam(lr=1e-6), loss=dice_coef_loss)

model_checkpoint_2 = ModelCheckpoint('weights_2.h5', monitor='val_loss', save_best_only=True)

model.fit(imgs_train, imgs_mask_train, batch_size=32, epochs=20, verbose=1, shuffle=True,
          validation_data=(imgs_val, imgs_mask_val), callbacks=[model_checkpoint_2])
```

Since the input to the U-net was resized, we need to scale the predicted masks back to the original size for the submission to kaggle leaderboard.

```python
image_rows = 420
image_cols = 580

from itertools import chain
    
def prep(img):
    img = img.astype('float32')
    img = (img > 0.5).astype(np.uint8)  # threshold
    img = resize(img, (image_rows, image_cols), preserve_range=True)
    return img


def run_length_enc(label):
    x = label.transpose().flatten()
    y = np.where(x > 0)[0]
    if len(y) < 10:  # consider as empty
        return ''
    z = np.where(np.diff(y) > 1)[0]
    start = np.insert(y[z+1], 0, y[0])
    end = np.append(y[z], y[-1])
    length = end - start
    res = [[s+1, l+1] for s, l in zip(list(start), list(length))]
    res = list(chain.from_iterable(res))
    return ' '.join([str(r) for r in res])
```

Write a Kaggle submission file.

```python
#imgs_mask_test = np.load('imgs_mask_test.npy')
argsort = np.argsort(test_imgs_id)
test_imgs_id = test_imgs_id[argsort]
imgs_mask_test = imgs_mask_test[argsort]

total = imgs_mask_test.shape[0]
ids = []
rles = []
for i in tqdm_notebook(range(total)):
    img = imgs_mask_test[i, :, :, 0]
    img = prep(img)
    rle = run_length_enc(img)

    rles.append(rle)
    ids.append(test_imgs_id[i])

first_row = 'img,pixels'
file_name = 'submission.csv'

with open(file_name, 'w+') as f:
    f.write(first_row + '\n')
    for i in tqdm_notebook(range(total)):
        s = str(ids[i]) + ',' + rles[i]
        f.write(s + '\n')
```
